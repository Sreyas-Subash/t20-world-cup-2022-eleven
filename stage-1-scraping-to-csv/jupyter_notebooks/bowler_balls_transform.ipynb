```python
df['runs']=df['runs'].map(lambda x : str(x).replace('1 run', '1'))
```

```python
df['wickets']=df['wickets'].map(lambda x : str(x).replace('2 s', '2'))
df['wickets']=df['wickets'].map(lambda x : str(x).replace('3 s', '3'))
```

```python
import ast
df['balls']=df['balls'].apply(ast.literal_eval)
df.loc[0,'balls']
```

```python
ball = []
def uni(y):
    for i in y:
        ball.append(i)
```

```python
import re

pattern = r'\d.*[a-z]$'

def dots(l):
    dots = 0
    for i in l:
        if i == '•':
            dots += 1
    return dots
    
def extras(l):
    extras = 0
    for i in l:
        if re.match(pattern, i):
            extras += int(i[0])

    return extras
            
            
```

```python
df['dot_balls'] = df['balls'].apply(dots)
df['extras'] = df['balls'].apply(extras)
```
