```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36',
    "Content-Type": "application/x-www-form-urlencoded"}
```

```python
player = []

for id in match_id_link:
    
    match_id = id
    match_url = match_id_link[id][0]
    # match_url = match_id_link['T20I # 1823'][0]
    
    match_html_text = requests.get(match_url,headers=headers).text
    match_soup = BeautifulSoup(match_html_text, 'lxml')
    
    innings_soup = match_soup.find_all('div', class_="ds-rounded-lg ds-mt-2")

    for innings in innings_soup:
        team = innings.find('span', class_="ds-text-title-xs ds-font-bold ds-text-typo").text
        
        soup = innings.find('table', class_="ds-w-full ds-table ds-table-md ds-table-auto ci-scorecard-table").find('tbody')
        each_row = soup.find_all('tr', class_="")

        for rows in each_row:
            
            each_value = rows.find_all('td')

            if (each_value[0].name == 'td') and (each_value[0].text == 'TOTAL'):
                break
            player_dict = {}

            player_dict['team'] = team.split('(')[0].strip()
            player_dict['player'] = each_value[0].text.split('\xa0')[0].strip()
            http = 'https://www.espncricinfo.com'
            player_dict['link'] = http + each_value[0].a['href']

            player.append(player_dict)

        did_not_bat_soup = soup.find('tr', class_='!ds-border-b-0')
        if did_not_bat_soup:
            did_not_bat = did_not_bat_soup.find_all('div', class_='ds-popper-wrapper ds-inline')
            for dnb_player in did_not_bat:
                player_dict = {}

                player_dict['team'] = team.split('(')[0].strip()
                player_dict['player'] = dnb_player.text.split('\xa0')[0].strip(', ')
                player_dict['link'] = http + dnb_player.a['href']

                player.append(player_dict)
```
