```python
import pandas as pd
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
df = pd.read_csv('bowling_data.csv')
```

```python
df.head()
```

```python

df['balls'] = df['overs'].map(lambda x: (int(str(x).split('.')[0]) * 6) + int((str(x).split('.')[1])) )
df
```

```python
df['bowling_team'].unique()
```

```python
df['bowling_team'] = df['bowling_team'].map(lambda x: str(x).replace('Netherlands  U.A.E.','Netherlands'))
df['bowling_team'] = df['bowling_team'].map(lambda x: str(x).replace('Sri Lanka  U.A.E.','Sri Lanka'))
df['bowling_team'] = df['bowling_team'].map(lambda x: str(x).replace('Namibia  U.A.E.','Namibia'))
```

```python
df
```

```python
for col in df:
    print(col, ':', df[col].nunique())
    print(df[col].unique())
```
